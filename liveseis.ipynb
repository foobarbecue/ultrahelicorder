Let's stream some data from Kilauea and plot it. [IRIS](https://www.iris.edu/hq/) is a big US national seismic organization which collects, archives, and distributes data, among other functions. 'HV' is the network code for [Hawaii Volcano Observatory](https://volcanoes.usgs.gov/observatories/hvo/). IRIS provides live streams in a format called seedlink, and we use obspy's easyseedlink to consume that feed.

```
from obspy.clients.seedlink.easyseedlink import create_client

def handle_trace(trace):
    trace.plot()

client = create_client('rtserve.iris.washington.edu', on_data=handle_trace)
client.select_stream('HV','WOOD','EHZ')
client.run()
```
